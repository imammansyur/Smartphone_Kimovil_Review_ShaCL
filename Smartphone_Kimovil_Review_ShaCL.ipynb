## SHACL

Kami membuat shape dari masing-masing kelas yang ada. 
Aturan yang kami terapkan adalah sebagai berikut:
* penggunaan "sh:maxCount 1" untuk semua property kecuali pada aggregateRating karena product dapat mempunyai lebih dari 1 aggregateRating
* tipe data yang ditentukan menyesuaikan dari yang telah ditentukan di vocabulary



```
shapes_graph = """
@prefix ex: <https://example.org/>.
@prefix rdfs:<http://www.w3.org/2000/01/rdf-schema#>.
@prefix rdf:<http://www.w3.org/1999/02/22-rdf-syntax-ns#>.
@prefix cci: <http://cookingbigdata.com/linkeddata/ccinstances/>.
@prefix sh:     <http://www.w3.org/ns/shacl#> .
@prefix xsd:    <http://www.w3.org/2001/XMLSchema#> .
@prefix schema: <http://schema.org/> .

schema:ProductShape
    a sh:NodeShape ;
    sh:targetClass schema:Product ;
    sh:property [
        sh:path schema:brand ;
        sh:maxCount 1 ;
        sh:class schema:Brand ;
    ] ;
    sh:property [
        sh:path schema:offers ;
        sh:maxCount 1 ;
        sh:class schema:Offer;
    ] ;

    sh:property [
        sh:path schema:url ;
        sh:maxCount 1 ;
        sh:datatype schema:URL ;
    ] ;

    sh:property [
        sh:path schema:aggregateRating ;
        sh:minCount 1 ;
        sh:class schema:AggregateRating ;
    ] .

schema:BrandShape
    a sh:NodeShape ;
    sh:targetClass schema:Brand ;
    sh:property [
        sh:path schema:name ;
        sh:maxCount 1 ;
        sh:datatype schema:Text;
    ] .

schema:OfferShape
    a sh:NodeShape ;
    sh:targetClass schema:Offer ;
    sh:property [
        sh:path schema:priceCurrency ;
        sh:maxCount 1 ;
        sh:datatype schema:Text;
    ] ;

    sh:property [
        sh:path schema:price ;
        sh:maxCount 1 ;
        sh:datatype schema:Number;
    ] .

schema:AggregateRatingShape
a sh:NodeShape ;
sh:targetClass schema:AggregateRating ;
sh:property [
    sh:path schema:ratingValue ;
    sh:maxCount 1 ;
    sh:datatype schema:Number;
    ] .

cci:InstanceShape
    a sh:NodeShape ;
    sh:targetClass cci:Instance ;
    sh:property [
        sh:path cci:hasRAM ;
        sh:maxCount 1 ;
        sh:class cci:ram ;
    ] ;
    sh:property [
        sh:path cci:hasStorage ;
        sh:maxCount 1 ;
        sh:class cci:storage ;
    ] ;
    sh:property [
        sh:path cci:hasCPU ;
        sh:maxCount 1 ;
        sh:class cci:cpu ;
    ] .

cci:ramShape
    a sh:NodeShape ;
    sh:targetClass cci:ram ;
    sh:property [
        sh:path cci:ram_size ;
        sh:maxCount 1 ;
        sh:datatype xsd:integer ;
    ] .

cci:storageShape
    a sh:NodeShape ;
    sh:targetClass cci:storage ;
    sh:property [
        sh:path cci:storage_size ;
        sh:maxCount 1 ;
        sh:datatype xsd:integer ;
    ] .

cci:cpuShape
    a sh:NodeShape ;
    sh:targetClass cci:cpu ;
    sh:property [
        sh:path cci:cpu_model ;
        sh:maxCount 1 ;
        sh:datatype xsd:string ;
    ] .
"""
```

```
import pyshacl

results = pyshacl.validate(
    data_graph,
    shacl_graph=shapes_graph,
    data_graph_format="ttl",
    shacl_graph_format="ttl",
    inference="rdfs",
    debug=True,
    serialize_report_graph="ttl",
    )

conforms, report_graph, report_text = results

print("conforms", conforms)
```
